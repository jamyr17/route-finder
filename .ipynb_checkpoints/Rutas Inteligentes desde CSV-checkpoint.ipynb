```python
# Imports
import csv 
import networkx as nx
import matplotlib.pyplot as plt
import heapq
import math
```

```python
# Constants
MINIMUM_NODES = 15
MINIMUM_EDGES = 40
```

```python
# Method to read nodes from Nodes.csv
def get_nodes(file_path):
    nodes = []
    
    try:
        with open(file_path, 'r') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader: 
                name, coordinate_x, coordinate_y = row 
                nodes.append({
                    'Name': name,
                    'X': float(coordinate_x),
                    'Y': float(coordinate_y)
                })

    # Taking in mind some possible errors
    except ValueError:
        print("An error ocurred parsing coordinates values from: " + file_path)
    except FileNotFoundError:
        print("An error ocurred because " + file_path + " was not found")
    except PermissionError:
        print("An error ocurred because required permissions for " + file_path + " are not satisfied")
    except Exception as e:
        print(f"Unexpected error: {str(e)}")

    finally:
        return nodes
```

```python
# Method to read edges from Edges.csv
def get_edges(file_path):
    edges = []
    
    try:
        with open(file_path, 'r') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader: 
                source, destination, weight = row 
                edges.append({
                    'Source': source,
                    'Destination': destination,
                    'Weight': float(weight)
                })

    # Taking in mind some possible errors
    except ValueError:
        print("An error ocurred parsing weight values from: " + file_path)
    except FileNotFoundError:
        print("An error ocurred because " + file_path + " was not found")
    except PermissionError:
        print("An error ocurred because required permissions for " + file_path + " are not satisfied")
    except Exception as e:
        print(f"Unexpected error: {str(e)}")

    finally:
        return edges
```

```python
# Method to add nodes recovered from .csv to graph
def add_nodes_to_graph(nodes, G):
    for node in nodes:
        G.add_node(node.get('Name'), coords=(node.get('X'), node.get('Y')))

    return G
```

```python
# Method to add edges recovered from .csv to graph
def add_edges_to_graph(edges, G):
    for edge in edges:
        G.add_edge(edge.get('Source'), edge.get('Destination'), weight=edge.get('Weight'))

    return G
```

```python
# Digraph's visualization
def visualize_digraph(G):
    plt.figure(figsize=(12, 12))
    
    # Optimized layout  
    pos = nx.spring_layout(
        G,
        k=0.4,          # Distance between nodes 
        iterations=100, 
        seed=25         
    )
    
    # Nodes configuration
    nx.draw_networkx_nodes(
        G,
        pos,
        node_size=250,      
        node_color='lightblue',
        edgecolors='gray',  
        linewidths=0.1,     
        alpha=0.9          
    )
    
    # Node labels configuration
    nx.draw_networkx_labels(
        G,
        pos,
        font_size=7,      
        font_family='sans-serif',
        font_color='black'
    )
    
    # Edges configuration
    nx.draw_networkx_edges(
        G,
        pos,
        width=0.2,        
        edge_color='gray',
        alpha=0.3         
    )
    
    # Edge labels (relations) configuration
    edge_labels = nx.get_edge_attributes(G, "weight")
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_size=7,      
        font_color='red',
        alpha=0.7       
    )
    
    # Final configuration
    plt.axis('off')       
    plt.tight_layout()    
    plt.show()
```

```python
# Method to ask user for source and destination
def ask_values(nodes):
    node_names = [node.get('Name') for node in nodes]
    
    # Welcome message
    print("\n" + "="*60)
    print(" WELCOME TO INTELLIGENT ROUTE FROM CSV ".center(60, '='))
    print("="*60)
    
    print("\nWe're here to help you find the shortest path between two points.")
    print("To do this, we need you to specify your starting point and destination.\n")
    
    # Display available nodes
    print("Available points are:")
    for name in node_names:
        print(f"{name}")
    
    while True:
        # Choosing source and validating its existence
        source = input("\nChoose source node: ")
        if source not in node_names:
            print("***The node you entered does not exist, please review the available options then try again***")
            continue

        # Choosing destination and validating its existence
        destination = input("\nChoose destination node: ")
        if destination not in node_names:
            print("***The node you entered does not exist, please review the available options then try again***")
            continue
        
        break

    # Summary message
    print("\n" + "="*60)
    print(f"Great! We'll find the best route from:")
    print(f"{source} -> {destination}") 
    print("="*60 + "\n")
    
    return source, destination
```

```python
# Method for heuristic
def heuristic_euclidean(node1, node2):
    return math.sqrt((node1['X'] - node2['X']) ** 2 + (node1['Y'] - node2['Y']) ** 2)
```

```python
# Method to search for node based on name
def get_node(name, nodes):
    for n in nodes:
        if n['Name'] == name:
            return n
    return None
```

```python
def algorithm(start, goal, nodes, edges):
    # Initialize cost and heuristic dictionaries
    g = {n['Name']: float('inf') for n in nodes}  # Actual cost from start to node
    f = {n['Name']: float('inf') for n in nodes}  # Estimated total cost (g + h)
    
    # g(n): actual cost from start to node n
    # h(n): heuristic estimating cost from n to goal
    # f(n) = g(n) + h(n): total estimated cost to minimize
    # start: name of starting node (e.g., "A")
    # goal: name of target node
    # nodes: list of nodes from Nodes.csv
    # edges: list of edges from Edges.csv
    # g: stores minimum cost from start to each node
    # f: estimates total cost (g + h) to goal

    came_from = {}  # Stores how we reached each node

    start_node = get_node(start, nodes)  # Find start node in nodes list
    goal_node = get_node(goal, nodes)   # Find goal node in nodes list

    g[start] = 0  # Actual cost to reach start node is 0
    f[start] = euclidean_heuristic(start_node, goal_node)  # Estimate cost from start to goal

    open_list = [(f[start], start)]
    heapq.heapify(open_list)  # Convert list to priority queue

    expanded = []  # Stores nodes removed from open_list

    while open_list:
        _, current = heapq.heappop(open_list)
        expanded.append(current)

        # -----------------------------------------------------------
        # Reconstruct and return path if goal is reached
        # -----------------------------------------------------------
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1], g[goal], expanded  # Return reversed path, cost, and expanded nodes
        # -----------------------------------------------------------

        # Explore possible paths
        for edge in edges:  # Find all edges from current node
            if edge['Source'] == current:
                neighbor = edge['Destination']  # Neighboring node that can be reached
                tentative_g = g[current] + edge['Weight']  # Actual cost to reach neighbor via current node

                if tentative_g < g[neighbor]:  # If new path is cheaper
                    came_from[neighbor] = current
                    g[neighbor] = tentative_g  # Update lowest cost
                    neighbor_node = get_node(neighbor, nodes)
                    f[neighbor] = tentative_g + euclidean_heuristic(neighbor_node, goal_node)
                    heapq.heappush(open_list, (f[neighbor], neighbor))  # Add neighbor to exploration list with f as priority

    return None, float('inf'), expanded  # No path found from start to goal
```

```python
def main():
    # Recoverying nodes and validating if program can continue or not 
    nodes = get_nodes('./Nodes.csv')
    if len(nodes) < MINIMUM_NODES:
        print("Minimun quantity of nodes (" + str(MINIMUM_NODES) + ") is not satisfied, so no further operations can be made")
        return

    # Recoverying edges and validating if program can continue or not 
    edges = get_edges('./Edges.csv')
    if len(edges) < MINIMUM_EDGES:
        print("Minimun quantity of edges (" + str(MINIMUM_EDGES) + ") is not satisfied, so no further operations can be made")
        return

    # As nodes and edges were recovered, now the digraph can be created 
    G = nx.DiGraph()
    G = add_nodes_to_graph(nodes, G)
    G = add_edges_to_graph(edges, G)
    visualize_digraph(G)

    # Asking for input values (source and destination)
    source, destination = ask_values(nodes)

    # Calling Algorithm A*
    path, cost, expanded = algorithm(source, destination, nodes, edges)

    if path:
        print("\nShortest path found:")
        print(" → ".join(path))
        print(f"Total cost: {cost}")
        print(f"Nodes expanded: {', '.join(expanded)}")
    else:
        print("No path found between selected nodes.")
main()
```
