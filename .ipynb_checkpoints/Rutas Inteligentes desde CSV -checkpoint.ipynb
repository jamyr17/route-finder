```python
# Imports
import csv 
import networkx as nx
import matplotlib.pyplot as plt
import heapq
import math
```

```python
# Constants
MINIMUM_NODES = 15
MINIMUM_EDGES = 40
```

```python
# Method to read nodes from Nodes.csv
def get_nodes(file_path):
    nodes = []
    
    try:
        with open(file_path, 'r') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader: 
                name, coordinate_x, coordinate_y = row 
                nodes.append({
                    'Name': name,
                    'X': float(coordinate_x),
                    'Y': float(coordinate_y)
                })

    # Taking in mind some possible errors
    except ValueError:
        print("An error ocurred parsing coordinates values from: " + file_path)
    except FileNotFoundError:
        print("An error ocurred because " + file_path + " was not found")
    except PermissionError:
        print("An error ocurred because required permissions for " + file_path + " are not satisfied")
    except Exception as e:
        print(f"Unexpected error: {str(e)}")

    finally:
        return nodes
```

```python
# Method to read edges from Edges.csv
def get_edges(file_path):
    edges = []
    
    try:
        with open(file_path, 'r') as file:
            reader = csv.reader(file)
            next(reader)
            for row in reader: 
                source, destination, weight = row 
                edges.append({
                    'Source': source,
                    'Destination': destination,
                    'Weight': float(weight)
                })

    # Taking in mind some possible errors
    except ValueError:
        print("An error ocurred parsing weight values from: " + file_path)
    except FileNotFoundError:
        print("An error ocurred because " + file_path + " was not found")
    except PermissionError:
        print("An error ocurred because required permissions for " + file_path + " are not satisfied")
    except Exception as e:
        print(f"Unexpected error: {str(e)}")

    finally:
        return edges
```

```python
# Method to add nodes recovered from .csv to graph
def add_nodes_to_graph(nodes, G):
    for node in nodes:
        G.add_node(node.get('Name'), coords=(node.get('X'), node.get('Y')))

    return G
```

```python
# Method to add edges recovered from .csv to graph
def add_edges_to_graph(edges, G):
    for edge in edges:
        G.add_edge(edge.get('Source'), edge.get('Destination'), weight=edge.get('Weight'))

    return G
```

```python
# Digraph's visualization
def visualize_digraph(G):
    plt.figure(figsize=(9, 9))
    
    # Optimized layout  
    pos = nx.spring_layout(
        G,
        k=0.4,          # Distance between nodes 
        iterations=100, 
        seed=25         
    )
    
    # Nodes configuration
    nx.draw_networkx_nodes(
        G,
        pos,
        node_size=100,      
        node_color='lightblue',
        edgecolors='gray',  
        linewidths=0.1,     
        alpha=0.9          
    )
    
    # Node labels configuration
    nx.draw_networkx_labels(
        G,
        pos,
        font_size=7,      
        font_family='sans-serif',
        font_color='black'
    )
    
    # Edges configuration
    nx.draw_networkx_edges(
        G,
        pos,
        width=0.2,        
        edge_color='gray',
        alpha=0.3         
    )
    
    # Edge labels (relations) configuration
    edge_labels = nx.get_edge_attributes(G, "weight")
    nx.draw_networkx_edge_labels(
        G,
        pos,
        edge_labels=edge_labels,
        font_size=6,      
        font_color='red',
        alpha=0.4       
    )
    
    # Final configuration
    plt.axis('off')       
    plt.tight_layout()    
    plt.show()
```

```python
# Method to ask user for source and destination
def ask_values(nodes):
    node_names = [node.get('Name') for node in nodes]
    
    # Welcome message
    print("\n" + "="*60)
    print(" WELCOME TO INTELLIGENT ROUTE FROM CSV ".center(60, '='))
    print("="*60)
    
    print("\nWe're here to help you find the shortest path between two points.")
    print("To do this, we need you to specify your starting point and destination.\n")
    
    # Display available nodes
    print("Available points are:")
    for name in node_names:
        print(f"{name}")
    
    while True:
        # Choosing source and validating its existence
        source = input("\nChoose source node: ")
        if source not in node_names:
            print("***The node you entered does not exist, please review the available options then try again***")
            continue

        # Choosing destination and validating its existence
        destination = input("Choose destination node: ")
        if destination not in node_names:
            print("***The node you entered does not exist, please review the available options then try again***")
            continue
        
        break

    # Summary message
    print("\n" + "="*60)
    print(f"Great!\nWe'll find the best route from:")
    print(f"{source} → {destination}") 
    print("="*60 + "\n")
    
    return source, destination
```

```python
# H1: Euclidean distance between current node and goal node
def heuristic_euclidean(node1, node2):
    return math.sqrt((node1['X'] - node2['X']) ** 2 + (node1['Y'] - node2['Y']) ** 2)
```

```python
# H2: Esimation of minimun remaining steps (1 in this case)
def heuristic_steps(n1, n2):
    return 1
```

```python
# Search for a node base on its name
def get_node(name, nodes):
    for n in nodes:
        if n['Name'] == name:
            return n
    return None
```

```python
# Algorithm A* that finds the shortest path from 'start' to 'goal'
# Parameters:
# - start initial node
# - goal goal node
# - nodes list of nodes
# - edges list of edges
# - heuristic: heuristic function to be used Euclidean or steps
def A(start, goal, nodes, edges, heuristic):
    g = {n['Name']: float('inf') for n in nodes}  # g(n) actual cost accumulated from start.
    f = {n['Name']: float('inf') for n in nodes}  # f(n) total estimate (g + h)

    came_from = {}  # Stores the previous node from which each node was reached

    # Gets the start and end nodes
    start_node = get_node(start, nodes)
    goal_node = get_node(goal, nodes)

    # The cost of reaching the start is 0
    g[start] = 0
    # Calculate initial estimate f(n) = g(n) + h(n)
    f[start] = heuristic(start_node, goal_node)

    # Initialize the priority queue with the start node
    open_list = [(f[start], start)]
    heapq.heapify(open_list)  #  Convert the list to a heap (priority queue)

    expanded = []  # Lista de nodos que fueron expandidos

    # A* algorithm loop
    while open_list:
        # Extract the node with smallest f(n).
        _, current = heapq.heappop(open_list)
        expanded.append(current)   # Save the expanded node

        # If it reaches the destination, then rebuilds the path.
        if current == goal:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1], g[goal], expanded  # Returns path, total cost, and expanded nodes

        # Traverses the outgoing edges of the current node
        for edge in edges:
            if edge['Source'] == current:
                neighbor = edge['Destination']  # Neighbor node to which it has edge
                cost = edge['Weight']  # Cost of the step between current and neighbor

                # Calculates g tentatively if that path is taken
                tentative_g = g[current] + cost

                # If a lower route to the neighbor is found
                if tentative_g < g[neighbor]:
                    came_from[neighbor] = current  # saved where arrived
                    g[neighbor] = tentative_g      # Update g with the new cost
                    neighbor_node = get_node(neighbor, nodes)  # lookup neighbor coordinates

                    # f(n) = g(n) + h(n)  with the chosen heuristic
                    f[neighbor] = tentative_g + heuristic(neighbor_node, goal_node)

                    # Add the neighbor to the queue with its priority f
                    heapq.heappush(open_list, (f[neighbor], neighbor))

    # If no path is found for goal it terminates and returns None
    return None, float('inf'), expanded
```

```python
def main():
    # Recoverying nodes and validating if program can continue or not 
    nodes = get_nodes('./Nodes.csv')
    if len(nodes) < MINIMUM_NODES:
        print("Minimun quantity of nodes (" + str(MINIMUM_NODES) + ") is not satisfied, so no further operations can be made")
        return

    # Recoverying edges and validating if program can continue or not 
    edges = get_edges('./Edges.csv')
    if len(edges) < MINIMUM_EDGES:
        print("Minimun quantity of edges (" + str(MINIMUM_EDGES) + ") is not satisfied, so no further operations can be made")
        return

    # As nodes and edges were recovered, now the digraph can be created 
    G = nx.DiGraph()
    G = add_nodes_to_graph(nodes, G)
    G = add_edges_to_graph(edges, G)
    visualize_digraph(G)

    # Asking for input values (source and destination)
    source, destination = ask_values(nodes)

    # -------------------------------------------------------------------------
    # Operations with 2 heuristics

    # A* with Euclidean Heuristic
    path1, cost1, expanded1 = A(source, destination, nodes, edges, heuristic_euclidean)

    # A* with Euclidean Steps
    path2, cost2, expanded2 = A(source, destination, nodes, edges, heuristic_steps)

    # Results for Euclidean Heuristic 
    print("\nResults using Euclidean Heuristic:")
    if path1:
        print("Route:", " → ".join(path1))
        print("Total cost:", cost1)
        print("Expanded nodes:", len(expanded1))
    else:
        print("No possible route was found")
        
    #  Results for Step-Based Heuristic 
    print("\nResults using Step-Based Heuristic")
    if path2:
        print("Route:", " → ".join(path2))
        print("Total cost:", cost2)
        print("Expanded nodes:", len(expanded2))
    else:
        print("No possible route was found")
        
    #  Heuristics Comparison
    print("\nComparison:")
    print(f"Euclidean: cost={cost1}, expanded nodes={len(expanded1)}")
    print(f"Step-Based: cost={cost2}, expanded nodes={len(expanded2)}")

main()
```
